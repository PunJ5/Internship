```python
from bs4 import BeautifulSoup
import requests
```

```python
page= requests.get('https://internshala.com/fresher-jobs')
page
```

```python
page.content
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
titles=soup.find_all('div',class_="heading_4_5 profile")
titles
```

```python
job_titles=[]

for i in titles:
    job_titles.append(i.text)
    
job_titles
```

```python
len(job_titles)
```

```python
companies=soup.find_all('a',class_= 'link_display_like_text')
companies
```

```python
company_names=[]

for i in companies:
    company_names.append(i.text.strip())
    
company_names
```

```python
len(company_names)
```

```python
locations=soup.find_all('p',id ="location_names")
locations
```

```python
location_list=[]
for i in locations:
    if i.text is None:
        location_list.append("no data")
    else:
        location_list.append(i.text.replace("\n",''))
location_list    
```

```python
len(location_list)
```

```python
startdate=soup.find_all('div',class_='other_detail_item')
startdate
```

```python
len(startdate)
```

```python
stipend=soup.find_all('span',class_='stipend')
stipend
```

```python
applyby=soup.find_all('div',class_='apply_by')
applyby
```
